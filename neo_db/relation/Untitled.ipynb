```python
f = open('./city2school.txt', 'w')
with open('./school_city.txt','r') as fs:
    lines = fs.readlines()
    for line in lines:
        try:
            school, city = line.strip().split()
            city = city[:2]
            f.write("{},{},{},{},{}\n".format(city, school, '包含', '城','大学'))
        except:
            print(line)
f.close()
```

```python
f = open('./city2river.txt', 'w')
with open('./river_city.txt', 'r') as fs:
    lines = fs.readlines()
    for line in lines:
        all_entity = line.strip().split('、')
        print(all_entity)
        for city in all_entity[1:]:
            f.write("{},{},{},{},{}\n".format(all_entity[0], city, '流经', '河流','城'))
f.close()
```

```python
all_river = []
with open('./city2river.txt') as f:
    lines = f.readlines()
    for line in lines:
        all_river.append(line.strip().split(',')[0])
```

```python
all_river = list(set(all_river))
```

```python
all_school =[]
with open('./city2school.txt') as f:
    lines = f.readlines()
    for line in lines:
        all_school.append(line.strip().split(',')[1])
```

```python
all_school = list(set(all_school))
```

```python
all_spot = []
with open('./city2spot.txt') as f:
    lines = f.readlines()
    for line in lines:
        all_spot.append(line.strip().split(',')[1])
        all_city.append(line.strip().split(',')[0])
```

```python
all_city = list(set(all_city))
all_spot = list(set(all_spot))
```

```python
entity2id ={}
json_data = {"data":[], 'links':[]}
```

```python
def add_data(datalist, idx, i):
    for data in datalist:
        if data == '宜宾':
            print(data)
        json_data['data'].append({'name':data, 'category':idx})
        entity2id[data] = i
        i = i+1
    return i
i = add_data(all_city, 0, 0)
i = add_data(all_school, 1, i)
i = add_data(all_spot, 2, i)
i = add_data(all_river, 3, i)
```

```python
print(i)
```

```python
def add_links(file):
    with open(file) as f:
        lines = f.readlines()
        for line in lines:
            line = line.strip().split(',')
            try:
                print(entity2id[line[0]])
                json_data['links'].append({"source":entity2id[line[0]], "target":entity2id[line[1]], 'value': line[2]})
            except:
                pass
add_links('./relation.txt')
```

```python
import json
json.dump(json_data,open('./relation.json', 'w'))
```
