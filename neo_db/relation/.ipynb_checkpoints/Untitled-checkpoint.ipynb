```python
f = open('city2school', 'w')
with open('./school_city.txt','r') as fs:
    lines = fs.readlines()
    for line in lines:
        school, city = line.strip().split()
        city = city[:3]
        print(school, city)
        f.write("{},{},{},{},{}\n".format(city, school, '包含', '城','大学'))
f.close()
```
